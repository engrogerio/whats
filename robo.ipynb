```python
# this should be read from a file
customer = """
    Nome: Rogério Vieira
    Endereco: Rua Rio Grande,690
    Numero: 4444
"""

welcome = """
Bem-vindo ao restaurante Rio Grande Rotisserie. 
Qual o seu pedido de hoje?
"""

bye = "Obrigado pelo seu pedido! Nosso tempo de entrega é de 30 a 50 min. Bom apetite!"

cart = """
pedido_atual: Vazio
"""

menu = """
pratos:
    1: Carne Assada = R$29,00
    2: Cação Ensopado = R$29,00
    3: Talharim com Almondegas = R$29,00
    4: Rondelli 4 queijos(Vegetariano) = R$29,00

bebidas:
    5: Suco de laranja = R$8,00
    6: Coca lata = R$4,00
    7: Coca lata diet = R$8,00
        
sobremesas:
    8: Gelatina
    9: Paçoca
"""

options = """
OPÇÕES:
    p: mostrar pedido
    m: mostrar menu do dia
    d: alterar prato
    f: finalizar a compra
    a: alterar dados de entrega
"""

nl = "___________________________"

message = [welcome,customer, nl, cart, nl, menu, nl, options, bye]
print(message)
```

```python
import yaml
import sys

class Robot:
    def __init__(self, menu, customer, options, cart, welcome):
        self.menu = yaml.load(menu, Loader=yaml.FullLoader)
        self.customer = yaml.load(customer, Loader=yaml.FullLoader)
        self.options = yaml.load(options, Loader=yaml.FullLoader)
        self.cart = yaml.load(cart, Loader=yaml.FullLoader)
        self.welcome = yaml.load(welcome, Loader=yaml.FullLoader)
        self.nl = yaml.load(nl, Loader=yaml.FullLoader)
        self.bye = yaml.load(bye, Loader=yaml.FullLoader)
        self.add_commands()
    
    def add_commands(self):
        commands = {}
        commands['p'] = self.show_cart
        commands['m'] = self.show_menu
        commands['f'] = self.show_bye
        commands['s'] = sys.exit
        self.commands = commands
    
    def render(self, message):
        if type(message) is dict:
            for id, item in message.items():

                if type(item) is not dict:
                    print(f'{id}:', item)
                else:
                    print(f'\n{id.upper()}:')

                if type(item) not in (list, dict):
                    pass
                else:
                    self.render(item)

        elif type(message) is list:
            for item in message:
                if type(item) not in (list, dict):
                    pass
                else:
                    render(item)
        else:
            print(message)       

    def show_linebreak(self):
        self.render(self.nl)
    
    def show_welcome(self):
        self.render(self.welcome)
    
    def show_customer(self):
        self.render(self.customer)
    
    def show_cart(self):
        self.render(self.cart)
    
    def show_menu(self):
        self.render(self.menu)

    def show_options(self):
        self.render(self.options)
    
    def show_bye(self):
        self.render(self.bye)
        
    def repl(self):
        nl = self.show_linebreak
        self.show_customer()
        nl()
        self.show_welcome()
        nl()
        self.show_cart()
        nl()
        self.show_menu()
        nl()
        self.show_options()
        entry = ''
        while entry != 's':
            entry = input('?').lower()
            self.show_linebreak()
            self.commands[entry]()
            self.show_options()
        self.show_bye()
        
        
```

```python

r = Robot(menu, customer, options, cart, welcome)
r.repl()
```
